```python
df=pd.read_csv("fifa_players.csv")
```

```python
df=df.drop_duplicates()
```

```python
df["full_name"]=df["full_name"].str.title()
```

```python
df=df.applymap(lambda x:x.strip()if isinstance(x,str)else x)
```

```python
df.dropna(inplace=True)
```
