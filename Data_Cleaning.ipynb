```python
import pandas as pd#import pandas
df = pd.read_csv("survey_practice.csv")#load the csv file
```

```python
df.head()#shows the 1st 5 rows
```

```python
df.head(10)
```

```python
df.info()
```

```python
df["Gender"]=df["Gender"].str.strip()
```

```python
df["Gender"]=df["Gender"].str.lower()
```

```python
df.head(10)
```

```python
df["County"]=df["County"].str.strip().str.title()# it removes spaces be4 and after the string and convert it into title format
```

```python
df.head(10)#shows 1st 10 rows
```

```python
df["Name"]=df["Name"].str.strip()#removes spaces before and after the string
```

```python
df.head(10)#show 10 rows
```

```python
df.isnull()#returns false when there is no null value and true when there is null values (show missing values)
```

```python
df.isnull().sum()   #adds all the null values in the columns
```

```python
df['Age'].fillna(0)#does not change the original dataset automatically unless you tell pandas to do s0
```

```python
df.dropna()
```

```python
df["Age"]=df['Age'].fillna(0)#replaces null with 0 in the original dataset
```

```python
df.head(10)
```

```python
df=df.drop_duplicates()
```

```python
df.head(10)
```

```python
df.to_csv("cleaned_data.csv",index=False)
```
